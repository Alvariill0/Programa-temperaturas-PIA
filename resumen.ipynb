## Trabajo con Bases de datos MySQL

####  - Conexión

```python
import mysql.connector    

# Establecer la conexión a la base de datos
conexion = mysql.connector.connect(
    host="localhost",
    user="root",
    password="**********",
    database="temperaturas")
```

### - Insertar 

```python
cursor = conexion.cursor()
consulta_fronteras = "INSERT INTO fronteras (idpais,cca3_frontera) VALUES(%s,%s)" 
valores_fronteras = ("id_pais", "frontera")
cursor.execute(consulta_fronteras, valores_fronteras)

#confirmar los cambios
conexion.commit()

# Cerrar el cursor y la conexión
cursor.close()
conexion.close()
```
